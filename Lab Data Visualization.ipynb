**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

**Box Plots**

Plot a box plot of Age.


```python
query = """
SELECT Age
FROM main
WHERE Age IS NOT NULL
"""
df_age = pd.read_sql_query(query, conn)

age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}

df_age['AgeNumeric'] = df_age['Age'].map(age_map)
df_age = df_age.dropna(subset=['AgeNumeric'])

plt.figure(figsize=(8, 5))
sns.boxplot(x=df_age['AgeNumeric'])
plt.title('Box Plot of Respondents\' Age')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
query = """
SELECT Age, WorkExp
FROM main
WHERE Age IS NOT NULL AND WorkExp IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70,
    'Prefer not to say': None
}

df['AgeNumeric'] = df['Age'].map(age_map)
df = df.dropna(subset=['AgeNumeric', 'WorkExp'])

plt.figure(figsize=(8, 6))
sns.scatterplot(x='AgeNumeric', y='WorkExp', data=df, alpha=0.6)
plt.title('Scatter Plot of Age vs. Work Experience')
plt.xlabel('Age')
plt.ylabel('Work Experience (years)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
query = """
SELECT TimeSearching, Frustration, Age
FROM main
WHERE TimeSearching IS NOT NULL AND Frustration IS NOT NULL AND Age IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

df['FrustrationCount'] = df['Frustration'].apply(lambda x: len(str(x).split(';')) if pd.notnull(x) else 0)

age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}
df['AgeNumeric'] = df['Age'].map(age_map)
df = df.dropna(subset=['AgeNumeric'])

# Bubble plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df,
    x='TimeSearching',
    y='FrustrationCount',
    size='AgeNumeric',
    sizes=(20, 400),
    alpha=0.5,
    legend=False
)

plt.title('Time Spent Searching vs Number of Frustration Sources\n(Bubble Size = Age)')
plt.xlabel('Time Spent Searching')
plt.ylabel('Number of Frustration Factors')
plt.grid(True)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
query = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

dbs_exploded = df['DatabaseWantToWorkWith'].str.split(';').explode().str.strip()

top5_dbs = dbs_exploded.value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(
    top5_dbs.values,
    labels=top5_dbs.index,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Top 5 Databases Respondents Want to Work With Next Year')
plt.axis('equal')
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
query = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age = '25-34 years old'
AND TimeSearching IS NOT NULL
AND TimeAnswering IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

time_order = {
    "Less than 15 minutes a day": 1,
    "15-30 minutes a day": 2,
    "30-60 minutes a day": 3,
    "60-120 minutes a day": 4,
    "More than 120 minutes a day": 5
}

df['TimeSearching_num'] = df['TimeSearching'].map(time_order)
df['TimeAnswering_num'] = df['TimeAnswering'].map(time_order)

median_searching = df['TimeSearching_num'].median()
median_answering = df['TimeAnswering_num'].median()

labels = ['Age 30–35 (≈25–34 group)']
searching = [median_searching]
answering = [median_answering]

plt.figure(figsize=(8, 6))
plt.bar(labels, searching, label='Time Searching')
plt.bar(labels, answering, bottom=searching, label='Time Answering')

plt.title('Stacked Bar Chart of Median TimeSearching and TimeAnswering (Age 30–35)')
plt.ylabel('Median Time (ordinal scale)')
plt.legend()
plt.tight_layout()
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
query = """
SELECT MainBranch FROM main
WHERE MainBranch IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

branch_counts = df['MainBranch'].value_counts()

plt.figure(figsize=(10, 6))
branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of Respondents by Main Branch')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.tight_layout()
plt.show()
```
