### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
# Column names
print("Column names:")
print(df.columns.tolist())

# Data types and non-null counts
print("\nData types and non-null values:")
print(df.info())

# Dataset shape
print(f"\nDataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Missing values
print("Missing values in columns:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# For not numeric values - filling with mode
for col in df.select_dtypes(include=['object', 'category']):
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

# Numeric values - filling with median
for col in df.select_dtypes(include=['int64', 'float64']):
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# If missing values >50%:
threshold = len(df) * 0.5
df.dropna(axis=1, thresh=threshold, inplace=True)

# Drop rows
df.dropna(inplace=True)

# Check
print("\nMissing values:", df.isnull().sum().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
key_columns = ['Employment', 'JobSat', 'YearsCodePro']

for col in key_columns:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts(dropna=False))
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['JobSat'], fill=True, bw_adjust=0.8)
plt.title('Job Satisfaction Distribution (KDE Plot)')
plt.xlabel('Job Satisfaction Score (0–10)')
plt.ylabel('Density')
plt.grid(True)
# Interpretation below the plot
plt.figtext(0.5, -0.1, 
            "Most responses are around score 7, indicating overall positive job satisfaction.", 
            wrap=True, horizontalalignment='center', fontsize=10)
plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Making a list
def get_language_counts(series):
    lang_series = series.dropna().str.split(';')
    lang_list = [lang.strip() for sublist in lang_series for lang in sublist]
    return pd.Series(lang_list).value_counts()

# Frequency
worked_counts = get_language_counts(df['LanguageHaveWorkedWith'])
want_counts = get_language_counts(df['LanguageWantToWorkWith'])

# Merge and fill missing values
lang_compare = pd.concat([worked_counts, want_counts], axis=1, keys=['WorkedWith', 'WantToWorkWith']).fillna(0)

# 10 most common languages
top_langs = lang_compare.sum(axis=1).sort_values(ascending=False).head(10)
lang_compare_top = lang_compare.loc[top_langs.index]

# Grouped bar chart
lang_compare_top.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Programming Languages: Worked With vs Want To Work With')
plt.xlabel('Programming Language')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# 1. Grouping
remote_region_counts = df.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')

# Top 10 countries
top_countries = df['Country'].value_counts().head(10).index

# Filtering for top10 country
filtered_data = remote_region_counts[remote_region_counts['Country'].isin(top_countries)]

# Pivot for heatmap
heatmap_data = filtered_data.pivot(index='Country', columns='RemoteWork', values='Count').fillna(0)

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='Blues')
plt.title('Remote Work Distribution by Country')
plt.xlabel('Remote Work')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
df['YearsCodePro_clean'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
}).astype(float)

# Pearson
correlation = df[['YearsCodePro_clean', 'JobSat']].corr(method='pearson')
print(correlation)
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Simple mapping
employment_map = {
    'Employed, full-time': 'Full-time',
    'Employed, part-time': 'Part-time',
    'Independent contractor, freelancer, or self-employed': 'Self-employed',
    'Not employed, but looking for work': 'Unemployed',
    'Not employed, and not looking for work': 'Not looking',
    'Retired': 'Retired',
    'Student, full-time': 'Student',
    'Student, part-time': 'Student',
    'I prefer not to say': 'Other',
    'Other': 'Other'
}

df['EmploymentSimplified'] = df['Employment'].map(employment_map)

# Top5 category edlevel
top_ed_levels = df['EdLevel'].value_counts().head(5).index
filtered_df = df[df['EdLevel'].isin(top_ed_levels)]

# Cross tabla
ct = pd.crosstab(filtered_df['EdLevel'], filtered_df['EmploymentSimplified'])

# Stacked bar plot
fig, ax = plt.subplots(figsize=(16, 9))
ct.plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Employment Status by Education Level (Simplified)')
ax.set_xlabel('Education Level')
ax.set_ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
