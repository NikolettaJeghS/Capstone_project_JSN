### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
comp = df['ConvertedCompYearly'].dropna()

upper_pct99 = comp.quantile(0.99)

plt.figure(figsize=(10, 5))
plt.hist(comp[comp <= upper_pct99], bins=30)
plt.title('Distribution of Yearly Compensation (0–99th percentile)')
plt.xlabel('ConvertedCompYearly (USD)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

df_clean = df.dropna(subset=['AgeNumeric'])

plt.figure(figsize=(8, 6))
plt.boxplot(df_clean['AgeNumeric'], vert=False, patch_artist=True)
plt.title('Distribution of Age (Numeric)')
plt.xlabel('Age (years)')
plt.yticks([1], ['Age'])
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

df_clean = df.dropna(subset=['AgeNumeric', 'ConvertedCompYearly'])

plt.figure(figsize=(10, 6))
plt.scatter(df_clean['AgeNumeric'], df_clean['ConvertedCompYearly'], alpha=0.5)
plt.title('Age vs. Yearly Compensation')
plt.xlabel('Age (years)')
plt.ylabel('ConvertedCompYearly (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

df_clean = df.dropna(subset=['AgeNumeric', 'ConvertedCompYearly', 'JobSatPoints_6'])

plt.figure(figsize=(10, 6))
sizes = df_clean['AgeNumeric'] * 10  # scale factor for visibility
plt.scatter(
    df_clean['ConvertedCompYearly'],
    df_clean['JobSatPoints_6'],
    s=sizes,
    alpha=0.5,
    edgecolors='w',
    linewidth=0.5
)
plt.title('Compensation vs. Job Satisfaction (Bubble size = Age)')
plt.xlabel('ConvertedCompYearly (USD)')
plt.ylabel('JobSatPoints_6')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
role_counts = df['MainBranch'].value_counts().sort_values()

plt.figure(figsize=(10, 6))
role_counts.plot(kind='barh')
plt.title("Distribution of Respondents' Primary Roles")
plt.xlabel('Number of Respondents')
plt.ylabel('Primary Role')
plt.tight_layout()
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
langs = df['LanguageWantToWorkWith'].dropna().str.split(';')
exploded = langs.explode().str.strip()
top5 = exploded.value_counts().head(5)

plt.figure(figsize=(8, 6))
top5.plot(kind='bar')
plt.title('Top 5 Programming Languages Respondents Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
df_clean = df.dropna(subset=['Age', 'JobSatPoints_6', 'JobSatPoints_7']).copy()

age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]

median_by_age = (
    df_clean
      .groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']]
      .median()
      .reindex(age_order)
)

median_by_age.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)
plt.title('Median Job Satisfaction (Points 6 & 7) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction (Points)')
plt.xticks(rotation=45, ha='right')
plt.legend(['JobSatPoints_6', 'JobSatPoints_7'], title='Question')
plt.tight_layout()
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
dbs = df['DatabaseHaveWorkedWith'].dropna().str.split(';')
db_exploded = dbs.explode().str.strip()
db_counts = db_exploded.value_counts()

plt.figure(figsize=(10, 6))
db_counts.head(10).plot(kind='bar')
plt.title('Top 10 Databases Respondents Have Worked With')
plt.xlabel('Database')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
df_clean = df.dropna(subset=['Age', 'ConvertedCompYearly']).copy()

age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]

median_comp = (
    df_clean
      .groupby('Age')['ConvertedCompYearly']
      .median()
      .reindex(age_order)
)

median_df = median_comp.reset_index()
median_df.columns = ['Age Group', 'Median Compensation (USD)']
median_df.set_index('Age Group', inplace=True)

median_df.plot(kind='bar', figsize=(10, 6))
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median ConvertedCompYearly (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
country_counts = df['Country'].value_counts()

top_n = country_counts.head(10)
top_n['Other'] = country_counts.iloc[10:].sum()

plt.figure(figsize=(10, 6))
top_n.sort_values().plot(kind='barh')
plt.title('Number of Respondents by Country (Top 10 + Other)')
plt.xlabel('Number of Respondents')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
